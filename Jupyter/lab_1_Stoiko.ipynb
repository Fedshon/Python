## 3. Відповіді на контрольні питання

1. Система контролю версій (СКВ) - це програмне забезпечення, яке використовується для відстеження змін у файловій системі та управління історією версій файлів та проектів. Основна мета СКВ - це забезпечити можливість спільної роботи над проектом, відстежувати історію змін, повертатися до попередніх версій файлів та вирішувати конфлікти під час злиття змін від різних розробників. Принципова відмінність між SVN (Subversion) та Git:

    - SVN - централізована СКВ, де історія версій зберігається на сервері, а розробники отримують доступ до неї через сервер. Команди для роботи зі змінами виконуються на сервері.

    - Git - розподілена СКВ, де кожен розробник має локальний репозиторій з повною історією. Розробники можуть працювати офлайн та комітувати зміни локально, а потім синхронізувати їх з віддаленим репозиторієм. Гіт більше сприяє гнучкій роботі та гілкуванню.

2. Команда git add . додає всі зміни у вашому робочому каталозі до індексу.

3. Команда git commit створює коміт - фіксовану версію вашого проекту, яка включає в себе всі зміни, які ви додали до індексу за допомогою git add. Коміт містить повідомлення про коміт, яке зазвичай коротко описує, які зміни внесені у цьому коміті. Кожен коміт зберігається у історії репозиторію та має унікальний ідентифікатор.

4. Розміщення локальної копії Git-репозиторію проєкта на віддаленому, що знаходиться на GitHub, виконується за допомогою команди git push.
